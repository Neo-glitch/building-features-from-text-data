```python
with open("./datasets/biography.txt", "r") as file:
    file_contents = file.read()
    
print(file_contents)
```

```python
sentences = file_contents.split("\n")

n_gram_vectorizer = CountVectorizer(ngram_range=(2, 3))    # just bigrams and tri grams
```

```python
transformed_vector = n_gram_vectorizer.fit_transform(sentences)

vocabulary = n_gram_vectorizer.vocabulary_

vocabulary
```

```python
# gets freg of bi and tri gram xtracted from file
word_count = transformed_vector.toarray().sum(axis = 0)
word_count
```

```python
# gets all dict items in form of tuples that's bi gram or tri gram and their int id
vocabulary.items()
```

```python
# sort vocab , so most frequent ngram comes first
sorted_word_list = sorted([(word_count[i], n_gram) for n_gram, i in vocabulary.items()], reverse= True)

sorted_word_list
```

**Gen n-grams using nltk**

```python
train_text = ["A bird in hand is worth two in a bush.",
             "Good things come to those who grind.",
             "Those Airforces cost $120.",
             "There are other fishes in the sea.",
             "The ball is in your court.",
             "Mr. Lionel goes to Paris.",
             "Doogle Howser M.D."]
```

```python
from nltk import bigrams
from nltk import trigrams

from nltk.tokenize import word_tokenize
```

```python
# joins all input_sentence to form a single input text to pass to word tokenize
word_tokens = word_tokenize(" ".join(train_text))

word_tokens
```

```python
nltk_bigrams = bigrams(word_tokens)

list(nltk_bigrams)  # bigrams in tuple
```

```python
nltk_trigrams = trigrams(word_tokens)

list(nltk_trigrams)
```

```python
# n grams
from nltk import ngrams

fivegrams = ngrams(word_tokens, 5)

for grams in fivegrams:
    print(grams)
```

```python
# metrics

# helps determine whether occurence of bigram is meaningful
from nltk.collocations import BigramAssocMeasures
# looks through all ngrams in text as candidate collocations and checks if meaningful
from nltk.collocations import BigramCollocationFinder  
```

```python
word_tokens = word_tokenize(file_contents)
```

```python
bigram_measures = BigramAssocMeasures()

finder = BigramCollocationFinder.from_words(word_tokens)  # ret bigrams

finder.apply_freq_filter(3) # filter bigram with freq < 3 out
```

```python
# finds n best bigrams, here top 15, using raw_freq methid
matches = finder.nbest(bigram_measures.raw_freq, 15)  

matches
```
