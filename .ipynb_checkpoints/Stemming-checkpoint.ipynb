```python
# 6th

from nltk import word_tokenize
from nltk.stem import *
import pandas as pd
```

```python
input_tokens = ["overwhelming", "overwhelmingly",
               "husehd", "hush",
               "lying", "lied",
               "destabilize", "stability",
               "friendship", "friendships", "friendly", "friendless",
               "connect", "connections", "connected",
               "the", "these", "those",
               "motivational", "motivate", "motivating"]
```

```python
ps = PorterStemmer()
```

```python
# stem tokens
ps_stemmed_tokens = []

for token in input_tokens:
    ps_stemmed_tokens.append(ps.stem(token))
```

```python
stems_df = pd.DataFrame({
    "words": input_tokens,
    "porter stemmer": ps_stemmed_tokens
})

stems_df
```

```python
# another stemmer algo
ls = LancasterStemmer()

# stem tokens
ls_stemmed_tokens = []

for token in input_tokens:
    ls_stemmed_tokens.append(ls.stem(token))
```

```python
ls_stems_df = pd.DataFrame({
    "words": input_tokens,
    "porter stemmer": ls_stemmed_tokens
})

ls_stems_df
```

```python
# snowball stemmer stems in languages other than just english
print(SnowballStemmer.languages)
```

```python
ss = SnowballStemmer("english")

# stem tokens
ss_stemmed_tokens = []

for token in input_tokens:
    ss_stemmed_tokens.append(ss.stem(token))
```

```python
ss_stems_df = pd.DataFrame({
    "words": input_tokens,
    "porter stemmer": ss_stemmed_tokens
})

ss_stems_df
```

```python
# compares all stemmers
stems_df = pd.DataFrame({
    "words": input_tokens,
    "Porter stemmer": ps_stemmed_tokens,
    "Lancaster stemmer": ls_stemmed_tokens,
    "Snowball stemmer(english)": ss_stemmed_tokens
})

stems_df
```
